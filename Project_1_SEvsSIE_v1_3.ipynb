PROJECT_1

Проанализировать курс акций компании Шнайдер Электрик и их основного конкурента - Сименс.
Если корреляция положительная, то это реакция на изменение потребности рынка.
Если корреляция отрицательная, значит компании  борются между собой за крупных заказчиков ("отнимают заказчиков").

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

# Import Data
df = pd.read_csv("https://raw.githubusercontent.com/Strukov0107/TerVer_Ana_MatStat/master/Data_for_project/Data_SCH_SIEM.csv")
df.head(5)

# Prepare Data
x = df['<DATA>'].values.tolist()
y1 = df['<SCNEIDER>'].values.tolist()
y2 = df['<SIEMENS>'].values.tolist()
mycolors = ['tab:red', 'tab:blue', 'tab:green', 'tab:orange', 'tab:brown', 'tab:grey', 'tab:pink', 'tab:olive']      
columns = ['Siemens', 'Schneider Electric']

# Draw Plot 
fig, ax = plt.subplots(1, 1, figsize=(16,9), dpi= 80)
ax.fill_between(x, y1=y1, y2=0, label=columns[1], alpha=0.5, color=mycolors[1], linewidth=2)

ax.fill_between(x, y1=y2, y2=0, label=columns[0], alpha=0.5, color=mycolors[0], linewidth=2)

# Decorations
ax.set_title('History of stock quotes of Schneider Electric & SIEMENS company', fontsize=18)
ax.set(ylim=[0, 30])
ax.legend(loc='best', fontsize=12)
plt.xticks(x[::200], fontsize=10, horizontalalignment='center')
plt.yticks(np.arange(10.0, 160.0, 10.0), fontsize=10)
plt.xlim(-10, x[-1])

# Draw Tick lines  
for y in np.arange(10.0, 160.0, 10.0):    
    plt.hlines(y, xmin=0, xmax=len(x), colors='black', alpha=0.3, linestyles="--", lw=0.5)

# Lighten borders
plt.gca().spines["top"].set_alpha(0)
plt.gca().spines["bottom"].set_alpha(.3)
plt.gca().spines["right"].set_alpha(0)
plt.gca().spines["left"].set_alpha(.3)
plt.show()
```

```python


corr_matrix = df.loc[:, ['<SCNEIDER>', '<SIEMENS>']].corr()
corr_matrix
```

Коэффициент корреляции больше нуля (=0.68)
Значит изменение котировок не связано с переходом крупных клиентов от одной компании к другой, а зависит от общей привлекательности инвестиций в энергетическую отрасль.

```python
#  Найдём ковариацию этих двух величин:

n = len(y1)

C = np.cov(y1, y2)
print('ковариация равна ', C)
 
```

```python
#   Найдём коэффициент корреляции Пирсона: 

cp = np.corrcoef(y1, y2)

print('коэфф-т Пирсона = ', cp)
```

```python
cp2 = np.corrcoef(y1, y2)
cp2
```
